## Challenges in Optimization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def objective(x):
    return np.sin(x) + np.sin((10.0/3.0)*x)
```

```python
# INPUT RANGE
r_min, r_max = 2, 7.5
# Sample datapoints in the input range
inputs = np.arange(r_min, r_max, 0.1)
# compute targets
results = objective(inputs)

# PLOT - loss function + global minimum
plt.plot(inputs, results, color = 'green')
x_optima = 5.145735
plt.axvline(x=x_optima, ls='--', color='red', label = "Global minima")
plt.legend()
plt.title("U(x) = sin(x) + sin(10/3 * x)")
#plt.savefig("IKSA1.png")
plt.show()
```

```python
def derivative(x):
    return (10/3)*np.cos(10*x/3) + np.cos(x)
```

```python
plt.plot(inputs, results, color = 'green', label = "$U(x) = sin(x) + sin(\\frac{10x}{3})$")
x_optima = 5.145735
#plt.axvline(x=x_optima, ls='--', color='red', label = "Global minima")
plt.plot(x_optima, objective(x_optima), 'ro', label = 'Global minimum')
plt.legend(loc = 1, prop={'size': 11}, frameon=False, labelspacing = 0)
#plt.title("U(x) = sin(x) + sin(10/3 * x)")


alpha = 0.5 # learning rate
nb_max_iter = 10 # Nb max d'iteration
eps = 0.000001 # stop condition

x0 = 2.5 # start point
y0 = objective(x0)
plt.scatter(x0, objective(x0))

cond = eps + 10.0 # start with cond greater than eps (assumption)
nb_iter = 0 
tmp_y = y0
while nb_iter < nb_max_iter: #cond > eps and
    x0 = x0 - alpha * derivative(x0)
    y0 = objective(x0)
    nb_iter = nb_iter + 1
    cond = abs( tmp_y - y0 )
    tmp_y = y0
    #print (x0,y0,cond)
    plt.scatter(x0, y0, c = "gold")

plt.title("Large Step Size")

plt.savefig("large_step_size.png", bbox_inches='tight')
plt.show()
```

```python
def x3(x):
    return x**3

inputs1 = np.arange(-3,3, 0.01)
outputs1 = x3(inputs1)

def derivative1(x):
    return 3*(x**2)
```

```python
plt.plot(inputs1, outputs1, color = 'green')
#x_optima = 5.145735
#plt.axvline(x=x_optima, ls='--', color='red', label = "Global minima")
#plt.legend()
plt.title("U(x) = x**3")
#plt.savefig("IKSA1.png")
plt.show()
```

```python
plt.plot(inputs1, outputs1, color = 'green', label = "$U(x) = x^3$")
#x_optima = 5.145735
#plt.axvline(x=x_optima, ls='--', color='red', label = "Global minima")
plt.legend(frameon=False)

alpha = 0.01 # learning rate
nb_max_iter = 1000 # Nb max d'iteration
eps = 0.000001 # stop condition

x0 = 2.5 # start point
y0 = x3(x0)
plt.scatter(x0, x3(x0))

cond = eps + 10.0 # start with cond greater than eps (assumption)
nb_iter = 0 
tmp_y = y0
while nb_iter < nb_max_iter: #cond > eps and
    x0 = x0 - alpha * derivative1(x0)
    y0 = x3(x0)
    nb_iter = nb_iter + 1
    cond = abs( tmp_y - y0 )
    tmp_y = y0
    #print (x0,y0,cond)
    plt.scatter(x0, y0, c = 'gold')

plt.title("Saddle Point")

plt.savefig("saddle_point.png",  bbox_inches='tight')
plt.show()
```
